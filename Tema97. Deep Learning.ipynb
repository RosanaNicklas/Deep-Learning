# Puerta AND, y

```python
#Solo es 1 la salida cuando las 2 entradas están a 1
```

```python
import pandas as pd
df = pd.DataFrame({'Entrada_1': [0, 0, 1, 1],
'Entrada_2': [0, 1, 0, 1],
'Salida_12': [0, 0, 0, 1]})
df
```

```python
import numpy as np
```

```python
training_data = np.array([[0,0], [0,1], [1,0], [1,1]], "float32")
training_data
```

```python
target_data = np.array([[0], [0], [0], [1]], "float32")
target_data
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
training_data.shape # 4 filas 2 columnas (2 variables)
```

```python
len(training_data)
```

```python
target_data.shape # 1 columna
```

```python
model = Sequential()
# segun esto:
# input_dim es el numero de columnas
# 4 es el numero de filas, osea len(df)
# 1 es la salida, que es 0 o 1 pero es una sola salida.
model.add(Dense(4, input_dim = 2, activation="relu"))
model.add(Dense(1, activation = "sigmoid"))
```

```python
model.compile(optimizer='adam',
loss = 'mean_squared_error',
metrics=['binary_accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(training_data, target_data, epochs=10000, verbose=2)
```

```python
model.predict(training_data).round()
```

```python
predictions = (model.predict(training_data) > 0.5).astype("int32")
predictions
```

```python
y_pred = model.predict(training_data)
y_pred = np.round(y_pred).astype(int)
y_pred
```

```python
predictions = (model.predict(training_data) > 0.5).astype("int32")
predictions
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['binary_accuracy'], label = 'train')
plt.show()
```

```python
plt.plot(history.history['loss'], label = 'train')
plt.show()
```

```python
scores = model.evaluate(training_data, target_data)
print("\nAccuracy: %.2f%%" % (scores[1]*100))
```
